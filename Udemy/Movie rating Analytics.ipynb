```python
list1=[]
mylable=[]
sns.set_style('whitegrid')
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre==gen].Budget)
    mylable.append(gen)
fig,ax=plt.subplots() #fixing subplot ie only one subplot
fig.set_size_inches(11.7,8.27) #size of A4 Paper Note: The size can be adjusted by subplots onto data
h1=plt.hist(list1,bins=20,stacked=True,rwidth=1,label=mylable) 
plt.title("Movie Budget Distribution",fontsize=30,color='DarkBlue',fontname='console')
plt.ylabel("Number of movies",fontsize=25,color='Red')
plt.xlabel("Budgets",fontsize=25,color='Green')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(frameon=True,fancybox=True,shadow=True,framealpha=0.8,prop={'size':15}) # or: fontsize='xx-large' ,framealpha is greyish color in legend box
plt.show()
```
